```python
#function to loop through Foursquare AIP with location data
def getNearbyVenues(names, latitudes, longitudes, radius=2500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#get venue with Foursquare
socal_venues = getNearbyVenues(names=df_socal['city'],
                                 latitudes=df_socal['lat'], 
                                 longitudes=df_socal['lng'])
```

```python
#check the size of resulting dataframe:
print('size of the venue df:', socal_venues.shape)
print('There are {} uniques categories.'.format(len(socal_venues['Venue Category'].unique())))
socal_venues.head()
```

```python
# one hot encoding
socal_onehot = pd.get_dummies(socal_venues[['Venue Category']], prefix="", prefix_sep="")
t_columns = socal_onehot.columns
# add neighborhood column back to dataframe
socal_onehot['Neighborhood'] = socal_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [socal_onehot.columns[-1]] + list(socal_onehot.columns[:-1])
socal_onehot = socal_onehot[fixed_columns]
# group by Neighborhood
socal_grouped = socal_onehot.groupby('Neighborhood').mean().reset_index()

#Fundtion to put in pd df
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

#create df with top10
num_top_venues = 15
indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe with top10
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = socal_grouped['Neighborhood'] 

for ind in np.arange(socal_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(socal_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head() 
```

```python
# set number of clusters
kclusters = 8  #7
socal_grouped_clustering = socal_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(socal_grouped_clustering)
print(kmeans.labels_)
# add clustering labels
neighborhoods_venues_sorted1 = neighborhoods_venues_sorted
neighborhoods_venues_sorted1.insert(0, 'Cluster Labels', kmeans.labels_)

# merge to add latitude/longitude for each neighborhood
socal_merged = df_socal.join(neighborhoods_venues_sorted1.set_index('Neighborhood'), on='city')
print(socal_merged.shape)
```

```python
# create map
map_clusters = folium.Map(location=[33.9046, -118.3417], zoom_start=9)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(socal_merged['lat'], socal_merged['lng'], 
                                  socal_merged['city'], socal_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

Another Foursquare AIP search but this time with 'search' option, not 'explore' to restrict the search for 'Japanese Restaurant'

```python
#function to loop through Foursquare AIP with location data
def getNearbyVenues2(names, latitudes, longitudes):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&query={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            2500, 
            'Japanese Restaurant')
            
        # make the GET request
        results = requests.get(url).json()["response"]['venues']
        
        num_JR = 0
        num_R = 0
        for vv in results:
            num_R = num_R+1
            if not vv['categories']:
                pass
            elif(vv['categories'][0]['name'] == 'Japanese Restaurant'):
                num_JR = num_JR + 1
            elif(vv['categories'][0]['name'] == 'Sushi Restaurant'):
                num_JR = num_JR + 1
        
        JR_percentage = num_JR/num_R
        venues_list.append([(name,lat,lon,JR_percentage*100)])
        print(name,',', num_JR,',', "{:.1%}".format(JR_percentage))
        


    nearby_venues2 = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues2.columns = ['Neighborhood','Neighborhood Latitude', 'Neighborhood Longitude', 'Percentage Japanese Restaurant']
    
    return(nearby_venues2)
```
